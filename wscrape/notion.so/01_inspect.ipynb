## Decipher the Information in URLs

`https://www.indeed.com/jobs?q=python&l=new+york`

- **Base URL**
    - `https://www.indeed.com/jobs`
- **Query Parameters**
    - Start & Separators: `?`, `&`
    - Information: `q=python`, `l=new+york`

Every page is different. E.g., here is a weird thing that happens on this site: When typing a city name into the URL directly rather than running the search, the URL changes to `https://www.indeed.com/q-python-l-new-york-jobs.html`. The results are the same and one refers to the other, so it doesn't matter in this case.

Always spend time getting to know the page you want to scrape before you start writing code. It saves a lot of time and effort in the long run.
